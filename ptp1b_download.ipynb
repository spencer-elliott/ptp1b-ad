## Fetching an SRA run using SRA Toolkit

Processing the raw data in FASTQ format involves huge files, even when compressed!

The first time I ever did this type of analysis I had access to a HPC environment and used Slurm workload manager. But right now I only have the processing power of my laptop.



I'll start by downloading just one sample, from two mice in the PTP1B knockout group:
- SRR36691286 (C57BL/6; APP/PS1; PTP1B-/-)

This can easily be done with SRA Toolkit:

```python
%%bash

fasterq-dump -e 1 -p SRR36691286
# -e 1: Number of cores to use (will keep this small on my laptop)
# -p: Enables progress bar
```

Since this is paired-end sequencing fasterq-dump creates two output files - one has forward reads while the other has reverse reads.

These files are huge, so let's compress them into gzip right away:

```python
%%bash

gzip SRR36691286_1.fastq
gzip SRR36691286_2.fastq
```

Now we have some raw data in FASTQ format to practice analysis.
