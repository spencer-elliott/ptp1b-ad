## Quality control using FastQC

FastQC is an easy tool that will tell the quality of reads:

```python
%%bash

fastqc SRR36691286_1.fastq.gz SRR36691286_2.fastq.gz
```

Now we can check out the html report that it made.

Here are how some of the most important parts of quality inspection turned out:
- Per base sequence quality - Quality (Phred) scores look good across all the reads.
- Per sequence quality scores - Almost all reads are of the highest quality.
- Adapter content - 0 across the whole read length.
- Overrepresented sequences - There are no overrepresented sequences in either file.

While some flags are thrown up in other categories this seems to show we have good data to work with. Also from what I can tell, run 2 is usually a little lower quality.
