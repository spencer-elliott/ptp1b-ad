## Fetching an SRA run using SRA Toolkit

Processing the raw data in FASTQ format involves huge files, even when compressed!

The first time I ever did this type of analysis I had access to a HPC environment and used Slurm workload manager. But right now I only have the processing power of my laptop.



I'll start by downloading just one sample, pooled from two mice in the PTP1B knockout group:
- SRR36691286 (C57BL/6; APP/PS1; PTP1B-/-)

This can easily be done with SRA Toolkit, and then some quality control can be done with FastQC. 

To manage all the different tools I need to use, I will make Conda environments for each phase of the project. So the initial step is to make an environment with the tools for downloading reads and quality control:

```python
%%bash

conda create -n fastqc_env fastqc sra-tools -c bioconda -c conda-forge
# -n fastqc_env: Name for the new environment
# fastqc: Quality control for .fastq files
# sra-tools: Downloading SRA accessions (raw data)
# -c bioconda -c conda-forge: Channels where Conda can find these packages

conda activate 
```

Before getting started, I also need to make a directory to store the raw data:

```python
%%bash

mkdir -p data/fastq
# mkdir: Makes a new directory
# -p: Create parent directory ('data') alongside new directory ('fastq')
```

Now SRA Toolkit is ready to go and can be run in one command:

```python
%%bash

fasterq-dump -e 1 -p SRR36691286 -O data/fastq
# -e 1: Number of cores to use (will keep this small on my laptop)
# -p: Enables progress bar
# -O: Output directory
```

Since this is paired-end sequencing fasterq-dump creates two output files - one has forward reads while the other has reverse reads.

These files are huge, so let's compress them into gzip right away:

```python
%%bash

gzip data/fastq/SRR36691286_1.fastq
gzip data/fastq/SRR36691286_2.fastq
```

Now we have some raw data in FASTQ format to practice analysis.
